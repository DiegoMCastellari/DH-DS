# Creamos una función para modificar los colores de los pixels una vez clasificados

```python
def color_class(df):
    
    for i in range(len(df)):
        if df.iloc[i][3] == 1:
            df.iloc[i] = [56,168,0,0]
        elif df.iloc[i][3] == 2:
            df.iloc[i] = [0,197,255,0]
        elif df.iloc[i][3] == 3:
            df.iloc[i] = [255,255,0,0]
        elif df.iloc[i][3] == 4:
            df.iloc[i] = [230,0,0,0]
        elif df.iloc[i][3] == 5:
            df.iloc[i] = [0,0,0,0]
        elif df.iloc[i][3] == 6:
            df.iloc[i] = [255,170,0,0]
    
    return df.iloc[:,:-1]
```

```python
def color_class_apply(df):
    
    if df['clase'] == 1:
        df['R'] == 56
        df['G'] == 168
        df['B'] == 0
        
    elif df['clase'] == 2:
        df['R'] == 0
        df['G'] == 197
        df['B'] == 255
        
    elif df['clase'] == 3:
        df['R'] == 255
        df['G'] == 255
        df['B'] == 0
        
    elif df['clase'] == 4:
        df['R'] == 230
        df['G'] == 0
        df['B'] == 0
        
    elif df['clase'] == 5:
        df['R'] == 0
        df['G'] == 0
        df['B'] == 0
        
    elif df['clase']  == 6:
        df['R'] == 255
        df['G'] == 170
        df['B'] == 0
    
    return df.iloc[:,:-1]
```

```python
def color_class_index(df):

    df.loc[df.clase == 1, 'R'] = 56
    df.loc[df.clase == 1, 'G'] = 168
    df.loc[df.clase == 1, 'B'] = 0
        
    df.loc[df.clase == 2, 'R'] = 0
    df.loc[df.clase == 2, 'G'] = 197
    df.loc[df.clase == 2, 'B'] = 255
        
    df.loc[df.clase == 3, 'R'] = 255
    df.loc[df.clase == 3, 'G'] = 255
    df.loc[df.clase == 3, 'B'] = 0
        
    df.loc[df.clase == 4, 'R'] = 230
    df.loc[df.clase == 4, 'G'] = 0
    df.loc[df.clase == 4, 'B'] = 0
        
    df.loc[df.clase == 5, 'R'] = 0
    df.loc[df.clase == 5, 'G'] = 0
    df.loc[df.clase == 5, 'B'] = 0
        
    df.loc[df.clase == 6, 'R'] = 255
    df.loc[df.clase == 6, 'G'] = 170
    df.loc[df.clase == 6, 'B'] = 0
        
    return df.iloc[:,:-1]
```

# Clasificamos el DF de la imagen y modificamos el RGB

```python
df_cls = df_ic_981.copy()
im_cls = im_jpg.copy()
```

```python
# Clasificamos el DF de toda la imagen
clase = modelo.predict(df_cls)
```

```python
# incorporamos la clase al DF y nos quedamos con los campos originales
df_cls['clase'] = clase
df_cls_im = df_cls.loc[:,['R', 'G', 'B', 'clase']]
```

```python
df_cls.to_csv('DF_img_completa_981_clasificada_3.csv', sep=',', encoding='utf-8')
```

```python
# corremos la función anterior para cambiar los colores RGB del DF según la clase predicha
df_cls_im = color_class_index(df_cls_im)
```

# Reconstruimos la imagen

```python
# volvemos a generar la imagen a partir del DF
img_csf = [tuple(l) for l in df_cls_im.values]
```

```python
# modificamos la imagen original y la guardamos
im_cls.putdata(img_csf)
im_cls.save('classified_3.jpg')
```
