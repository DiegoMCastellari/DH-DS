```python
from PIL import Image
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
img_compl = Image.open('San Pedro - Misiones 1.jpg')
```

# DF con los datos de los pixels cercanos (d=1)

```python
d_img_compl = 1
pixel_list_9 = []            # lista donde se recopilan los datos del pixel y vecinos

for yy in range(0, img_compl.size[1],1):                  # se itera sobre filas y columnas    
        for xx in range(0, img_compl.size[0], 1):
            pixel_fila = []                           # lista donde se recopilan los datos de cada ventana
            pixel_pos = img_compl.getpixel(xy=(xx,yy))      # se indexa el pixel centrtal
            pixel_fila += pixel_pos                   # se colocan los datos en la lista
            
            for d1 in range(-d_img_compl, (d_img_compl+1), 1):                   # se itera sobre filas y columnas alrededor del pixel central 
                for d2 in range(-d_img_compl, (d_img_compl+1), 1):
                    if (d1 != 0) or (d2 != 0):                                              # se excluye el pixel central          
                        if (0 <xx+d1< img_compl.size[0]) and (0 <yy+d2< img_compl.size[1]):       # se limita al rango de la imagen         
                            pixel_around = img_compl.getpixel(xy=(xx+d1,yy+d2))       # se indexan los valores de los pixeles
                            pixel_fila += pixel_around                          # se adjuntan a la lista parcial

            #pixel_list_9.append(pixel_fila)        # se adjunta la fila a la lista general
            a = list(range(0,len(pixel_fila),3))
            b = list(np.array(a) + 1)
            c = list(np.array(a) + 2)
            
            # se separa la lista por color y transforma en array para poder aplicar los métodos de numpy
            pixel_fila_R = np.array(pixel_fila)[a]
            pixel_fila_G = np.array(pixel_fila)[b]
            pixel_fila_B = np.array(pixel_fila)[c]
                     
            listaRGB = []
            # se adjunta a lista la media y desvío de cada color
            listaRGB += (pixel_fila_R[0],pixel_fila_R.mean(), pixel_fila_R.std())        
            listaRGB += (pixel_fila_G[0],pixel_fila_G.mean(), pixel_fila_G.std())   
            listaRGB += (pixel_fila_B[0],pixel_fila_B.mean(), pixel_fila_B.std())  
            
            pixel_list_9.append(listaRGB)
```

```python
df_com_9 = pd.DataFrame(pixel_list_9)
df_com_9.head()
```

```python
df_com_9.columns = ['R', 'R9mean', 'R9std', 'G', 'G9mean', 'G9std', 'B', 'B9mean', 'B9std']
df_com_9.head()
```

```python
df_com_9 = df_com_9.loc[:,['R', 'G', 'B', 'R9mean', 'G9mean', 'B9mean', 'R9std', 'G9std', 'B9std']]
df_com_9.head()
```

```python
df_com_9.to_csv('DF_img_completa_9_2.csv', sep=',', encoding='utf-8', index=True)
```
